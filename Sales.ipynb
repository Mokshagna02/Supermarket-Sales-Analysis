```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
df['Hour'] = df.Time.apply(lambda x: x.split(':')[0])
df['Minutes'] = df.Time.apply(lambda x: x.split(':')[1])
```

```python
# Total sales by branch
sales_by_branch = df.groupby('Branch')['Total'].sum().sort_values(ascending=False)

print(sales_by_branch)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=sales_by_branch.index,y=sales_by_branch.values,palette='Greens_r')
plt.title('Sales in each Branch')
plt.ylabel('Total Sales')
plt.xlabel('Branch')
plt.show()
```

```python
# Total sales by city
sales_by_city = df.groupby('City')['Total'].sum().sort_values(ascending=False)

print(sales_by_city)
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=sales_by_city.index,y=sales_by_city.values,palette='Reds_r')
plt.title('Sales in each City')
plt.xlabel('city')
plt.ylabel('sales')
plt.show()
```

```python
sales_by_product=df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
sales_by_product
```

```python
# Adding a 'Day Name' column to the dataframe
df['Day name'] = df['Date'].dt.day_name()

# Analyzing sales by day of the week
sales_by_day = df.groupby('Day name')['Total'].sum().sort_values(ascending=False)

print(sales_by_day)
```

```python
sns.barplot(x=sales_by_day.index,y=sales_by_day.values,palette='viridis')
plt.title('Sales Per Day')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.show()
```
